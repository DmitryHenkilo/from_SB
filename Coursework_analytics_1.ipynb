```python
import numpy as np
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from datetime import date
```

```python
# Загружаем датасеты и избавляемся от мусора
df_courses = pd.read_csv('courses.csv', usecols = ['id','title','field'])
df_students = pd.read_csv('students.csv', usecols = ['id','city','birthday'])
df_course_contents = pd.read_csv('course_contents.csv')
df_progresses = pd.read_csv('progresses.csv')
df_progress_phases = pd.read_csv('progress_phases.csv')
```

```python
# Изменяем названия некоторых колонок в датасетах для их объединения
df_courses.rename(columns = {'id':'course_id'}, inplace = True)
df_progresses.rename(columns = {'id':'progress_id'}, inplace = True)
df_students.rename(columns = {'id':'student_id'}, inplace = True)
```

```python
# Создаем общий датасет на основе "progresses", так как в нем есть как id курса, так и id прогресса, что необходимо для дальнейшего объединения
df_all = df_progresses.merge(df_courses, on = 'course_id')
# Добавим данные из "course_contents"
df_all = df_all.merge(df_course_contents, on = 'course_id')
# Добавим данные из "df_progress_phases"
df_all = df_all.merge(df_progress_phases, on = (['module_number','lesson_number','progress_id']))
```

```python
# Приведение дат начала и окончания курса к типу "datetime64"
df_all['start_date'] = df_all['start_date'].astype('datetime64')
df_all['finish_date'] = df_all['finish_date'].astype('datetime64')
```

```python
# Подсчитаем общее количество курсов в датасете
num_of_courses = df_all['title'].unique().size
num_of_courses
```

```python
# Подсчитаем количество модулей на каждом курсе
num_of_modules = df_all.groupby('title')['module_title'].apply(lambda x: x.unique().size).sort_values(ascending = False)
num_of_modules
```

```python
# Подсчитаем количество уроков в каждом модуле на каждом курсе
num_of_lessons = df_all.groupby(['title', 'module_title'])['lesson_title'].apply(lambda x: x.unique().size)
num_of_lessons
```

```python
# Подсчитаем медианное количество уроков в модуле на каждом курсе
med_num_of_lessons = df_all.groupby(['title', 'module_title'])['lesson_title'].apply(lambda x: x.unique().size).groupby("title").median().sort_values(ascending = False)
med_num_of_lessons
```

```python
# Подсчитаем количество учеников на каждом курсе.
num_of_students = df_all.groupby(['title'])['student_id'].apply(lambda x: x.unique().size).sort_values(ascending = False)
num_of_students
```

```python
# Подсчитаем минимальный, максимальный, средний и медианный возраст студентов
# В отдельный датасет отберем всех студентов с известными датами рождения
df_students_age = df_students[df_students['birthday'].notnull()]
# В отдельную переменную запишем последнюю актуальную дату датасета и примем ее за дату отсчета
today = df_all['finish_date'].max()
# Добавим возраст студентов в готовый датасет
df_students_age['age'] = df_students_age['birthday'].apply(lambda x: today.year - date.fromisoformat(x).year)
# Удалим возраст, который мы считаем аномальным.
df_students_age = df_students_age.query("age >= 18 and age <= 75")
# Выведем минимальный, максимальный, средний, медианный возраст студентов.
min_age = df_students_age['age'].min()
max_age = df_students_age['age'].max()
mean_age = df_students_age['age'].mean()
median_age = df_students_age['age'].median()
print('Минимальный возраст студентов -', min_age,'\nМаксимальный возраст студентов -', max_age, '\nСредний возраст студентов -', round(mean_age), '\nМедианный возраст студентов -', round(median_age))
```

```python
# Подсчитаем минимальный, максимальный, средний и медианный возраст студентов на каждом курсе
ages_on_course = df_all.merge(df_students_age, on = 'student_id').groupby('title')['age'].agg([np.min, np.max, np.mean, np.median])
ages_on_course.sort_values(by = 'median', ascending = False)
```

```python
# Cтроим bar-chart, отражающий количество студентов на каждом курсе.
# За основу берем подготовленный датасет с количеством студентов на курсе из предыдущего задания.
fig, ax = plt.subplots(figsize = (20,10))
ax.set_title("Количество студентов на каждом курсе")
ax.bar(num_of_students.index, num_of_students.values)
ax.set_xlabel("Название курса")
ax.set_ylabel("Количество студентов")
plt.xticks(rotation = 90)
plt.show()
```

```python
# (p) Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю 
# (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом 
# по дате начала первой домашней работы.
df_stud_grow = df_all[df_all['is_homework'] == True].groupby(['title','student_id'])['start_date'].min().reset_index()
# Присваиваем дате начала прохождения модуля тип "datetime64"
df_stud_grow['start_date'] = df_stud_grow['start_date'].astype('datetime64[M]')
# Создадим список дат
date_list = pd.date_range(df_stud_grow['start_date'].min(), df_stud_grow['start_date'].max(), freq='M').strftime('%Y-%m').to_list()
```

```python
# Присваиваем дате начала прохождения модуля тип "datetime64"
df_stud_grow['start_date'] = df_stud_grow['start_date'].astype('datetime64[M]')
# Создадим список дат
date_list = pd.date_range(df_stud_grow['start_date'].min(), df_stud_grow['start_date'].max(), freq='M').strftime('%Y-%m').to_list()
```

```python
# В подготовленный датасет добавим даты в виде колонок и при совпадении с двтой начала курса запишем "1" в ячейку.
for i in date_list:
    df_stud_grow.loc[df_stud_grow['start_date']==i, i] = 1
df_stud_grow
```

```python
# Готовый датасет переведем в более понятный вид
df_stud_grow = df_stud_grow.groupby(['title', 'student_id']).sum().stack().reset_index().rename(columns={'level_2': 'start_date', 0: 'mark'})
# Присваиваем дате начала прохождения модуля тип "datetime64"
df_stud_grow['start_date'] = df_stud_grow['start_date'].astype('datetime64[D]')
df_stud_grow.info()
```

```python
# Сгруппируем и аггрегируем датасет так, чтобы получить датасет, упорядоченый по названию курса и датам с числом вновь прибывших студентов за текущий месяц
df_stud_grow = df_stud_grow.groupby(['title','start_date']).agg(stud_count = ('mark', np.sum)).reset_index()
df_stud_grow.head()
```

```python
# Пройдем циклом по датасету для получения общего количества студентов на каждом курсе за каждый месяц с учетом нового пополнения
for i in df_stud_grow['title'].unique():
    stud = df_stud_grow.loc[df_stud_grow['title']==i, 'stud_count'].keys()
    for k in stud:
        df_stud_grow.loc[k, 'global_stud'] = df_stud_grow.loc[stud[0]:k, 'stud_count'].sum()
df_stud_grow
```

```python
# На основании первого пункта построить line-graph с приростом студентов в каждом месяце для каждого курса. 15 графиков. 
# Графики должны иметь заголовки, оси должны быть подписаны. Ticks нужно развернуть так, чтобы они были читаемы.
```

```python
# Создадим список названия курсов.
title_list = df_stud_grow['title'].unique()
# Построим графики под каждый курс из списка
fig, ax = plt.subplots(15,1, figsize = (15, 50))
for i, cours in enumerate(title_list):
    grow_df = df_stud_grow[df_stud_grow['title']==cours]
    a = ax[i]
    a.plot(grow_df['start_date'], grow_df['global_stud'])
    a.set_xlabel('')
    a.set_ylabel('К-во студентов')
    a.set_title(cours)
plt.suptitle('Графики распределения новых студентов по месяцам', y = 0.89)
plt.show()
```

```python
# Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, 
# приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), 
# к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).
```

```python
# Создаем датасет, в котором сгруппируем данные по каждому студенту на каждом курсе по дате начала и окончания выполнения домашнего задания
df_conv = df_all.loc[df_all['is_homework']==True, ['field','title','module_number','module_title','start_date','finish_date','student_id']].groupby(['field','title','module_number','module_title','student_id']).agg({'start_date':np.min, 'finish_date':np.max})
# Преобразуем даты в формат "datetime64[D]"
df_conv['start_date'] = df_conv['start_date'].astype('datetime64[D]')
df_conv['finish_date'] = df_conv['finish_date'].astype('datetime64[D]')
# Добавим в датасет метки закончил ли студент домашнее задание
df_conv['start_count'] = df_conv['start_date'].notnull()
df_conv['finish_count'] = df_conv['finish_date'].notnull()
df_conv = df_conv.groupby(['field','title','module_number','module_title']).agg({'start_count': np.sum, 'finish_count': np.sum, 'start_date': np.min, 'finish_date': np.max}).reset_index()
df_conv.head()
# Сферу, к которой относится курс и номер модуля мы добавили для реализации одного из следующих заданий
```

```python
# Снова используем список курсов, созданный ранее и циклом пройдем по каждому модулю, записывая в "start_count" количество 
# студентов начавших выполнение задания данного модуля, в "finish_count" количество студентов завершивших выполнение задания 
# данного модуля и записывая в основной датасет "df_conv" результаты с подсчетом конверсии текущего модуля.
for title in title_list:
    new = df_conv.loc[df_conv['title'] == title, 'module_number'].items()
    for i, module_number in new:
        start_count = df_conv.loc[i, 'start_count']
        finish_count = df_conv.loc[(df_conv['title']==title) & (df_conv['module_number'] < module_number), 'finish_count'].min()
        df_conv.loc[i, 'conversion'] = start_count / finish_count if finish_count > 0 else 0
```

Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

```python
# Создадим датасет, в который вынесем все записи, где дата начала выполнения урока раньше даты окончания
df_final = df_all[df_all['start_date'] < df_all['finish_date'] ].copy()
```

```python
# В отдельный датасет вынесем данные по началу и дате окончания урока и сгруппируем их по курсу и студенту
students_backlog = df_final.groupby(['title', 'student_id']).agg({'start_date': np.min, 'finish_date': np.max}).reset_index()
```

```python
# Посчитаем сколько дней прошло с момента завершения последнего урока используя переменную "today" из первого задания(в этой 
# переменной мы храним дату последней записи в общем датасете)
students_backlog['backlog_days'] = (today - students_backlog['finish_date']).dt.days
```

```python
# Создадим датасет  с общим количеством уроков в модуле
total_lessons = df_final.groupby(['title', 'module_title']).agg({'lesson_title': lambda x: x.unique().size}).groupby(['title']).agg(lessons=('lesson_title', np.sum)).reset_index()
```

```python
# Создадим датасет с количеством уроков, пройденных на данный момент
now_lesson = df_final.groupby(['title', 'student_id']).agg(lessons_done=('lesson_title', np.size)).reset_index()
```

```python
# Объединим полученные датасеты
lesson_metric = now_lesson.merge(total_lessons)
```

```python
# Добавим количество процентов, пройденных заданий
lesson_metric['done_percent'] = (lesson_metric['lessons_done']*100 / lesson_metric['lessons']).round(2)
```

```python
# Объединим датасеты с метрикой уроков и задержкой от учебного процесса
metric = lesson_metric.merge(students_backlog, on = (['title', 'student_id']))
```

```python
# В отдельный датасет вынесем тех студентов, которые закончили обучение
metric['group'] = metric['done_percent'].apply(lambda x: 'Done' if x >= 100 else 0)
done = metric[metric['group'] == 'Done']
done.groupby('title')['student_id'].count().sort_values()
```

```python
# Сформируем датасет из оставшихся студентов
uno = metric[metric['group'] == 0]
del uno['group']
uno['group'] = uno['backlog_days'].apply(lambda x: 'In_process' if x <= 30 else 0)
```

```python
# В отдельный датасет вынесем тех студентов, которые находятся в процессе обучения
in_process = uno[uno['group'] == 'In_process']
in_process.groupby('title')['student_id'].count().sort_values()
```

```python
# Оставшихся студентов распределим на тех, кто имеет проблемы в обучении и "забросивших учебу"
dos = uno[uno['group'] == 0]
del dos['group']
dos['group'] = dos['backlog_days'].apply(lambda x: 'Have_difficulties' if 30< x <= 180 else 'Dropped')
```

```python
# В отдельный датасет вынесем тех студентов, которые имеют проблемы в обучении
have_difficulties = dos[dos['group'] == 'Have_difficulties']
have_difficulties.groupby('title')['student_id'].count().sort_values()
```

```python
# В отдельный датасет вынесем тех студентов, которые бросили обучение
dropped = dos[dos['group'] == 'Dropped']
dropped.groupby('title')['student_id'].count().sort_values()
```

```python
all_studs = dos.append(in_process)
all_studs = all_studs.append(done)
all_in_pie = all_studs.set_index('student_id')['group'].value_counts()
fig, ax = plt.subplots(figsize = (10,10))
ax.pie(all_in_pie, labels = all_in_pie.index, data = all_in_pie.values, autopct='%1.1f%%', shadow = True, explode = (0.1, 0.1, 0.1, 0.1), startangle=90)
ax.set_title('Диаграмма соотношения студентов завершивших, бросивших, испытывающих трудности и продолжающих обучение')
plt.show()
```
